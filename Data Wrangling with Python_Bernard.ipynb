
## 5. Solution Implementation

Here we investigate the questions that would help craft our recommendations.

### 5.a) Questions

```
# 1. Which bus companies that had the highest breakdowns?
# ---
#
Highest_Bus_breakdown = Bus_breakdown.groupby(['BUS_COMPANY_NAME']).count()
# Sort to get bus company with highest breakdowns
#---
# YOUR CODE GOES BELOW
Highest_Bus_breakdown.sort_values (by='BREAKDOWN_OR_RUNNING_LATE', ascending=0).head(1)
```

```
# 2. What were the top 3 reasons for bus delays?
# ---
#
Bus_breakdown_delay = Bus_breakdown.groupby(['REASON']).count()[['HOW_LONG_DELAYED']].reset_index()

# Sort to get the most frequent reason
# ---
# YOUR CODE GOES BELOW
Bus_breakdown_delay.sort_values(by='HOW_LONG_DELAYED', ascending = 0).head(3)
```

```
# 3. How many students were in the buses when they broke down? 
# ---
#
Bus_breakdown.groupby(['NUMBER_OF_STUDENTS_ON_THE_BUS']).count()[['BUSBREAKDOWN_ID']].reset_index()
```

```
# 4. Which were most frequent reasons for bus breakdowns?
# ---
#
breakdown_reasons = Bus_breakdown[Bus_breakdown.BREAKDOWN_OR_RUNNING_LATE == 'breakdown'].groupby(['REASON']).count()
breakdown_reasons
# Sort to get most frequent reasons
# ---
# YOUR CODE GOES BELOW
breakdown_reasons.sort_values(by='BUSBREAKDOWN_ID', ascending = 0)
```

```
# 5. What were the most frequent reasons for the bus running late? 
# ---
#
reasons = Bus_breakdown.groupby(['REASON','HOW_LONG_DELAYED']).count()
reasons
# Get the records with running late reasons and sort to get most frequent reasons
# ---
# YOUR CODE GOES BELOW
reasons.sort_values (by='REASON', ascending=0)
```

```
# 6. What was the average delay time for each reason type?
# ---
# 
avg_delay = Bus_breakdown.groupby('REASON').mean().reset_index()
avg_delay
# Get the records with reasons and how long on average a delay took then sort 
# ---
# YOUR CODE GOES BELOW
#
avg_delay.sort_values (by='HOW_LONG_DELAYED', ascending=0)
```

### 5.b) Recommendations

From the above analysis, below are our recommendations:

1. g.v.c. ltd had the highest bus breakdowns and thus should service their buses to avoid breakdowns.
2. Heavy traffic, other and mechanical problems were the major reasons for bus delays hence bus companies should monitor and take routes with lesser traffic and service their buses to avoid delays caused by mechanical problems.
3. In most of the instances the buses were occupied whenever they had breakdowns and hence bus companies should make arrangements for backup buses to pick students and drop them whenever breakdowns happen.
4. Mechanical problems,flat tire and other were the most frequent reasons for bus breakdowns thus bus companies should ensure that their buses are always in good service and have well treaded tires
5. Buses that won't start were the main reasons for bus delays hence bus companies should ensure proper service is done around that area.
